```python
class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None
```

```python
import matplotlib.pyplot as plt

def plot_tree(root, x=0, y=0, parent_x=None, parent_y=None, level=0, spacing=1.5):
    if not root:
        return

    if parent_x is not None and parent_y is not None:
        plt.plot([parent_x, x], [parent_y, y], '-k', lw=2)

    plt.plot(x, y, 'ok', ms=15)
    plt.text(x, y, str(root.val), va='center', ha='center', color='white', fontsize=12)

    left_spacing = right_spacing = spacing

    if root.left:
        left_spacing = spacing / 2
        plot_tree(root.left, x=x - left_spacing, y=y - 1, parent_x=x, parent_y=y, level=level + 1, spacing=left_spacing)

    if root.right:
        right_spacing = spacing / 2
        plot_tree(root.right, x=x + right_spacing, y=y - 1, parent_x=x, parent_y=y, level=level + 1, spacing=right_spacing)

    if level == 0:
        plt.axis('off')
        plt.show()
```

```python
root = Node(7)
root.left = Node(3)
root.right = Node(12)
root.left.left = Node(99)
root.left.right = Node(8)
root.right.left = Node(5)
root.right.right = Node(6)
```

```python
plt.figure(figsize=(10,10))
plot_tree(root)
plt.axis('off')
plt.show()
```

```python
def printMaxLevel(root):
    if not root:
        return
    if not root.left and not root.right:
        print(root.val)
    elif not root.right or root.left.val > root.right.val:
        print(root.val)
        printMaxLevel(root.left)
    else:
        print(root.val)
        printMaxLevel(root.right)
```

```python
def insert(root, val):
    if root is None:
        root = Node(val)
    elif val < root.val:
        root.left = insert(root.left, val)
    else:
        root.right = insert(root.right, val)
    return root
```

```python
# 檢查二元樹是否建立成功
def inorder(root):
    if root is not None:
        inorder(root.left)
        print(root.val)
        inorder(root.right)
```

```python
# 按照先左後右順序插入給定序列建立二元樹
root_2 = None
values = [7, 3, 12, 99, 8, 5, 6]
for val in values:
    root_2 = insert(root_2, val)

inorder(root_2)
```

```python
plt.figure(figsize=(8,8))
plot_tree(root_2)
plt.axis('off')
plt.show()
```

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

def insertLevelOrder(arr, root, i, n):
    if i < n:
        if arr[i] is None:
            return None
        temp = Node(arr[i])
        root = temp
        root.left = insertLevelOrder(arr, root.left, 2*i+1, n)
        root.right = insertLevelOrder(arr, root.right, 2*i+2, n)
    return root

def buildTree(arr):
    n = len(arr)
    root = None
    root = insertLevelOrder(arr, root, 0, n)
    return root

arr = [7, 3, 12, 99, 8, 5, 6]
root_22 = buildTree(arr)
```

```python
plt.figure(figsize=(8,8))
plot_tree(root_22)
plt.axis('off')
plt.show()
```
